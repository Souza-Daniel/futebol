```python
import pandas as pd
```

```python
df = pd.read_csv('mundo_transfermarkt_competicoes_brasileirao_serie_a.csv', sep=',')
anos_campeonato = df.drop_duplicates(subset='ano_campeonato')['ano_campeonato']
dados_ano = {'Ano':anos_campeonato}
df_ano = pd.DataFrame(dados_ano)

df_ano.to_csv('Anos_campeonato.csv', index=False)
```

```python
times = df.drop_duplicates(subset='time_mandante')['time_mandante'].reset_index(drop=True)
dados_time = {'Time': times,
              'Estado': ''}

df_times = pd.DataFrame(dados_time)
df_times.insert(0, 'id', df_times.index + 1)
df_times.to_csv('Times.csv', index=False)

df_times
```
